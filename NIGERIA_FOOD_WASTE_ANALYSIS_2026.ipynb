Cleaning and projecting 2024 NBS data to 2026 values using compounded food inflation rates

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Historical Data (NBS Nov 2024)
data = {
    'Food_Item': ['Rice (Local)', 'Beans (Brown)', 'Yam Tuber', 'Garri (White)', 'Beef (Boneless)'],
    'Price_Nov_2024': [1100, 950, 1200, 800, 3500]
}
df = pd.DataFrame(data)

# 2. Projection Logic (The "Professional" Step)
# Applying estimated inflation to bridge the gap to April 2026
months_gap = 17
monthly_inflation = 0.025 # 2.5% monthly estimate
df['Price_April_2026'] = df['Price_Nov_2024'] * ((1 + monthly_inflation) ** months_gap)

# 3. Waste Calculation (UNEP Standard: 113kg per year = ~9.4kg/month)
monthly_waste_kg = 9.4
df['Monthly_Loss_Naira'] = df['Price_April_2026'] * monthly_waste_kg

# 4. Visualization for your Portfolio
plt.figure(figsize=(10, 6))
plt.bar(df['Food_Item'], df['Monthly_Loss_Naira'], color='forestgreen')
plt.title('Monthly Economic Loss Per Citizen Due to Food Waste (2026 Est.)')
plt.ylabel('Value in Naira (₦)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

print("--- ANALYSIS COMPLETE ---")
print(df[['Food_Item', 'Price_April_2026', 'Monthly_Loss_Naira']].round(2))
plt.show()
```
